```python
import pandas as pd
df = pd.read_csv("/data/ephemeral/home/hackathon/data/melb_data.csv")

df.head()
```

```python
# 데이터 타입별 컬럼 구분
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
boolean_columns = df.select_dtypes(include=['bool']).columns.tolist()
datetime_columns = df.select_dtypes(include=['datetime64']).columns.tolist()
other_columns = df.select_dtypes(exclude=['int64', 'float64', 'object', 'category', 'bool', 'datetime64']).columns.tolist()

# 결과 출력
print("범주형 데이터 컬럼:", categorical_columns)
print("수치형 데이터 컬럼:", numerical_columns)
print("Boolean 데이터 컬럼:", boolean_columns)
print("Datetime 데이터 컬럼:", datetime_columns)
print("기타 데이터 컬럼:", other_columns)
```

```python
# 특정 열의 결측치 비율 계산 함수
def calculate_missing_percentage(df, columns):
    """
    데이터프레임에서 특정 열의 결측치 비율을 계산합니다.
    
    Args:
        df (pd.DataFrame): 데이터프레임
        columns (list): 결측치를 계산할 열의 리스트
        
    Returns:
        dict: 열 이름과 해당 열의 결측치 비율(%)
    """
    return {col: df[col].isnull().mean() * 100 for col in columns}

# 'BuildingArea'와 'YearBuilt' 열의 결측치 비율 계산
missing_columns = ['BuildingArea', 'YearBuilt']
missing_percentages = calculate_missing_percentage(df, missing_columns)

# 결과 출력
for column, percentage in missing_percentages.items():
    print(f"'{column}' column has {percentage:.2f}% missing values.")
```
